```python
import cv2
import matplotlib.pyplot as pyplot
```

```python
img=cv2.imread(r'D:\DL Projects\face_detection\Snapchat-1320919387.jpg')
img
```

```python
import matplotlib.pyplot as plt
plt.imshow(img)
```

```python
cv2.imshow('IMAGE',img)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
conv_img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(conv_img)
```

```python
face_cascade=cv2.CascadeClassifier(r'D:\DL Projects\face_detection\haarcascade_frontalface_default.xml')
face_cascade
```

```python
img_gray=cv2.cvtColor(conv_img,cv2.COLOR_RGB2GRAY)
img_gray
```

```python
face=face_cascade.detectMultiScale(img_gray,minNeighbors=14)
face
```

```python
len(face)
```

```python
for (x,y,w,h) in face:
    cv2.rectangle(conv_img,(x,y),(x+w,y+h),(0,0,255),4)
```

```python
video=cv2.VideoCapture(r'D:\DL Projects\face_detection\video.mp4')
while True:
    success,frame=video.read()
    gray=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    faces=face_cascade.detectMultiScale(gray,minNeighbors=10)
    for (x,y,w,h) in faces:
        cv2.rectangle(frame,(x,y),(x+w,y+h),(0,122,232),3)
    cv2.imshow("VIDEO",frame)
    if cv2.waitKey(1) & 0XFF==ord('q'):
        break
video.release()
cv2.destroyAllWindows()
```
